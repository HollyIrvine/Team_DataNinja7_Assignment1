```python
%%bigquery --project supple-design-465411-c1
                                              #Profit increase from 2019 to 2022 by Category
SELECT
  t1.category,
  SUM(
    CASE
      WHEN EXTRACT(YEAR FROM t1.order_date) = 2019 THEN t1.profit
      ELSE 0
    END) AS profit_2019,
  SUM(
    CASE
      WHEN EXTRACT(YEAR FROM t1.order_date) = 2022 THEN t1.profit
      ELSE 0
    END) AS profit_2022,
  (SUM(
    CASE
      WHEN EXTRACT(YEAR FROM t1.order_date) = 2022 THEN t1.profit
      ELSE 0
    END) - SUM(
    CASE
      WHEN EXTRACT(YEAR FROM t1.order_date) = 2019 THEN t1.profit
      ELSE 0
    END)) / SUM(
    CASE
      WHEN EXTRACT(YEAR FROM t1.order_date) = 2019 THEN t1.profit
      ELSE 0
    END) * 100 AS profit_increase_percent
FROM
  `supple-design-465411-c1`.Lab1_Eda3.Super_Store2019_2022 AS t1
WHERE
  EXTRACT(YEAR FROM t1.order_date) IN (2019, 2022)
GROUP BY t1.category
HAVING SUM(
  CASE
    WHEN EXTRACT(YEAR FROM t1.order_date) = 2019 THEN t1.profit
    ELSE 0
  END) > 0
ORDER BY t1.category;

```

```python
# Operations_analysts

#1.1 Regional Performance Summary (by Region)
#This query calculates total sales, total profit, profit margin, and order count per region.

%%bigquery --project supple-design-465411-c1

SELECT
    Region,
    SUM(Sales) AS Total_Sales,
    SUM(Profit) AS Total_Profit,
    -- Calculate Profit Margin: (Total Profit / Total Sales) * 100
    (SUM(Profit) * 100.0 / NULLIF(SUM(Sales), 0)) AS Profit_Margin_Percent,
    COUNT(DISTINCT Order_ID) AS Total_Orders
FROM
    `supple-design-465411-c1`.Lab1_Eda3.Super_Store2019_2022
GROUP BY
    Region
ORDER BY
    Total_Sales DESC;
```

```python
 # Operations_analysts

# 1.2 State-Level Performance (Top 10 by Sales)
# This query details sales, profit, and profit margin for each state,
# ordered by total sales to identify top-performing states.

%%bigquery --project supple-design-465411-c1
SELECT
    state, -- Changed from "State" to "state"
    SUM(sales) AS Total_Sales,
    SUM(profit) AS Total_Profit,
    (SUM(profit) * 100.0 / NULLIF(SUM(sales), 0)) AS Profit_Margin_Percent,
    COUNT(DISTINCT order_id) AS Total_Orders
FROM
    `supple-design-465411-c1`.Lab1_Eda3.Super_Store2019_2022
GROUP BY
    state
ORDER BY
    Total_Sales DESC
LIMIT 10; -- Adjust LIMIT as needed for more/ fewer states
```

```python
# Operations_analysts

#1.3 State-Level Performance (Bottom 10 by Profit Margin)
#This query identifies states with the lowest profit margins,
#which might indicate areas for consolidation or operational review.

%%bigquery --project supple-design-465411-c1
SELECT
    state,
    SUM(sales) AS Total_Sales,
    SUM(profit) AS Total_Profit,
    (SUM(profit) * 100.0 / NULLIF(SUM(sales), 0)) AS Profit_Margin_Percent,
    COUNT(DISTINCT order_id) AS Total_Orders
FROM
    `supple-design-465411-c1`.Lab1_Eda3.Super_Store2019_2022
GROUP BY
    state
HAVING
    SUM(sales) > 0 -- Exclude states with no sales to avoid division by zero issues
ORDER BY
    Profit_Margin_Percent ASC
LIMIT 10; -- Adjust LIMIT as needed
```

```python
# Operations_analysts

#2.3 Performance by Product Category and Sub-Category
#This helps identify which product lines are contributing most to sales and profit,
#or which might be underperforming.


%%bigquery --project supple-design-465411-c1

SELECT
    category, -- Changed from "Category" to "category"
    subcategory, -- Changed from "Sub-Category" to "subcategory"
    SUM(sales) AS Total_Sales,
    SUM(profit) AS Total_Profit,
    (SUM(profit) * 100.0 / NULLIF(SUM(sales), 0)) AS Profit_Margin_Percent,
    COUNT(DISTINCT order_id) AS Total_Orders
FROM
    `supple-design-465411-c1`.Lab1_Eda3.Super_Store2019_2022
GROUP BY
    category,
    subcategory
ORDER BY
    Total_Profit DESC;
```

```python
# Operations_analysts

#FOCUS QUESTION 3: Where should we expand or consolidate?
#Required Analysis: Market opportunity identification


#3.1 States with High Sales but Low Profit (Potential for Consolidation/Efficiency Improvement)
#suggesting issues with costs, discounts, or operational inefficiencies.


%%bigquery --project supple-design-465411-c1

SELECT
    state,
    SUM(sales) AS Total_Sales,
    SUM(profit) AS Total_Profit,
    (SUM(profit) * 100.0 / NULLIF(SUM(sales), 0)) AS Profit_Margin_Percent
FROM
    `supple-design-465411-c1`.Lab1_Eda3.Super_Store2019_2022
GROUP BY
    state
HAVING
    SUM(sales) > 10000 -- Set a reasonable sales threshold to focus on significant markets
    AND (SUM(profit) * 100.0 / NULLIF(SUM(sales), 0)) < 5 -- Example: Profit margin less than 5%
ORDER BY
    Profit_Margin_Percent ASC, Total_Sales DESC;
```

```python
# Operations_analysts

#3.2 States with High Profit and Good Sales (Potential for Expansion)
#These are the "star" areas where the business is performing well.
#Expansion efforts could yield good returns here.

%%bigquery --project supple-design-465411-c1

SELECT
    state,
    SUM(sales) AS Total_Sales,
    SUM(profit) AS Total_Profit,
    (SUM(profit) * 100.0 / NULLIF(SUM(sales), 0)) AS Profit_Margin_Percent
FROM
    `supple-design-465411-c1`.Lab1_Eda3.Super_Store2019_2022
GROUP BY
    state
HAVING
    SUM(sales) > 50000 -- Set a reasonable sales threshold for significant markets
    AND (SUM(profit) * 100.0 / NULLIF(SUM(sales), 0)) > 15 -- Example: Profit margin greater than 15%
ORDER BY
    Profit_Margin_Percent DESC, Total_Sales DESC;
```

```python

# Operations_analysts

#3.3 Identify Underperforming Product Categories/Sub-Categories in Specific Regions
#This query helps pinpoint specific product lines that are struggling in certain regions,
#indicating where consolidation or strategic changes might be needed.

%%bigquery --project supple-design-465411-c1


SELECT
    region,
    category,
    subcategory,
    SUM(sales) AS Total_Sales,
    SUM(profit) AS Total_Profit,
    (SUM(profit) * 100.0 / NULLIF(SUM(sales), 0)) AS Profit_Margin_Percent
FROM
    `supple-design-465411-c1`.Lab1_Eda3.Super_Store2019_2022
GROUP BY
    region,
    category,
    subcategory
HAVING
    SUM(sales) > 1000 -- Consider only categories with meaningful sales
    AND SUM(profit) < 0 -- Focus on categories that are losing money
ORDER BY
    Total_Profit ASC;
```
